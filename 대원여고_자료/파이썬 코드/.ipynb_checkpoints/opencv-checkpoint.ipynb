```python
import cv2 

camera = cv2.VideoCapture(0)

width  = camera.get( cv2.CAP_PROP_FRAME_WIDTH )
height = camera.get( cv2.CAP_PROP_FRAME_HEIGHT )
print( f'해상도 가로 {width} 세로 {height}' )

camera.set(cv2.CAP_PROP_FRAME_WIDTH, 1280)
camera.set(cv2.CAP_PROP_FRAME_HEIGHT, 720)

while True:
    
    ret, image = camera.read()
    
    image = cv2.flip(image, 1)
    
    #image = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
    
    print(image, type(image), image.shape)

    #cv.rectangle(img, 좌상단, 우하단, color, thickness)
    cv2.rectangle(image, (20,10), (100,200), (255,0,0), 2)

    #cv.circle(	img, center, radius, color, thickness )
    cv2.circle(image, (100,100), 100, (0,255,0), 2)

    #cv.line( img, 시점, 종점, color, thickness)
    cv2.line(image, (100,0), (0,200), (255,255,255), 2)

    #cv.putText(img, text, org, fontFace, fontScale, color, thickness)
    cv2.putText(image, "opencv", (10,20), cv2.FONT_HERSHEY_SIMPLEX, 1, (0,0,255), 2)
    
    cv2.imshow('image',image)
    
    if cv2.waitKey(1)==ord('q'):
        break

camera.release()
cv2.destroyAllWindows()
```
